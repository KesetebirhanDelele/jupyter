```python
import pandas as pd
```

```python
from summarytools import dfSummary
```

```python
df = pd.read_csv('C:/Users/keset/Downloads/Demo_sales_data.csv')
```

```python
dfSummary(df)
```

```python
df.info() # Data type and non-missing values shown well
```

```python
df.describe() # Summary is useful but distribution is not shown and this is for numerical variables only
```

```python
df.Segment.value_counts() # You can calculate frequency of categorical variables one by one
```

```python
df.groupby('Segment').count() # Count all columns by a column
```

```python
df.groupby(['Segment','City']).count() # Count two columns
```

```python
df.groupby(['Segment','City'])['City'].count() # Count two columns but display count for only one column
```

```python
df['freq']=df.groupby('Segment')['Segment'].transform('count') # add a frequency column for total count of column categories. May help to calculate relative frequency
df.head(2)
```
