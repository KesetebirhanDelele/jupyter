```python
import pandas as pd
```

```python
from summarytools import dfSummary, tabset

sales = pd.read_csv('./data/Demo_sales_data.csv')
employees = pd.read_csv('./data/employees.csv')
supermarket_sales = pd.read_csv('./data/supermarket_sales.csv')
staffing = pd.read_csv('./data/staffing.csv')

tabset({
    'sales': dfSummary(sales).render(),
    'employees': dfSummary(employees).render(),
    'supermarket_sales': dfSummary(supermarket_sales).render(),
    'staffing': dfSummary(staffing).render()   
})
```
